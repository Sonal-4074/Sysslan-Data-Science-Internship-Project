```python
import pandas as pd

# Load dataset
df = pd.read_csv("Dataset1.csv")


# Basic insights

total_trains = df["Train_No"].nunique()
total_stations = df["Station_Name"].nunique()

max_distance = df["Distance"].max()
avg_distance = df["Distance"].mean()
min_distance = df["Distance"].min()


print("Key Exploratory Insights")
print("-------------------------")

print("Total number of trains:", total_trains)

print("Total number of stations:", total_stations)

print("Maximum route distance:", max_distance, "KM")

print("Average distance:", round(avg_distance,2), "KM")

print("Minimum route distance:", min_distance, "KM")
```
