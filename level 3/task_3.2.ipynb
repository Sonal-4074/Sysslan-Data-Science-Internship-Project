```python
import pandas as pd

# Load dataset
df = pd.read_csv("Dataset1.csv")


# Count how many trains pass through each station
station_traffic = df.groupby("Station_Name")["Train_No"].nunique().reset_index()

station_traffic.columns = ["Station_Name", "Train_Count"]


# Stations with highest train traffic
high_traffic = station_traffic.sort_values(
    "Train_Count",
    ascending=False
)

print("Stations with High Train Traffic:")
print(high_traffic.head(10))


# Stations with lowest train traffic
low_traffic = station_traffic.sort_values(
    "Train_Count",
    ascending=True
)

print("\nStations with Low Train Traffic:")
print(low_traffic.head(10))
```

```python
# Save output
high_traffic.to_csv("High_Train_Traffic.csv", index=False)
low_traffic.to_csv("Low_Train_Traffic.csv", index=False)
```
