```python
# Task 1.2 Train-wise Start and End Stations

import pandas as pd
import numpy as np

df = pd.read_csv("Dataset1.csv")

route_table = df.groupby('Train_No').agg(
    Start_Station=('Station_Name', 'first'),
    End_Station=('Station_Name', 'last')
).reset_index()

print(route_table.head())
```

```python
route_table.to_csv("Train_Route_Table.csv", index=False)
```
