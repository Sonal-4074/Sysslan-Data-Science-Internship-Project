```python
import pandas as pd

# Load dataset
df = pd.read_csv("Dataset1.csv")
```

```python
# Check missing/null values
print("Missing Values Count:")

missing_values = df.isnull().sum()

print(missing_values)
```

```python
# Total missing values in dataset
print("\nTotal Missing Values:")

print(df.isnull().sum().sum())
```

```python
# Check duplicate records
print("\nDuplicate Records:")

print(df.duplicated().sum())
```

```python
# Check data types for inconsistency
print("\nDataset Information:")

df.info()
```

```python
# Check inconsistent values in important columns

print("\nUnique Train Numbers:")
print(df['Train_No'].nunique())

print("\nUnique Route Numbers:")
print(df['Route_Number'].nunique())
```

```python
# Check negative distance values

negative_distance = df[df['Distance'] < 0]

print("\nNegative Distance Records:")
print(negative_distance)
```
