```python
# Task 1.3 Distance Statistics

import pandas as pd
import numpy as np

df = pd.read_csv("Dataset1.csv")

distance_stats = df['Distance'].describe()

print(distance_stats)
```

```python
# Number of stops per train
stops = df.groupby('Train_No')['Station_Name'].count().reset_index()

stops.columns = ['Train_No', 'Number_of_Stops']

print("\nNumber of Stops:")
print(stops)
```

```python
# Stops statistics
print("\nStops Statistics:")

print(stops['Number_of_Stops'].describe())
```
