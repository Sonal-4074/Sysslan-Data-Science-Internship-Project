## Task 4.4: Patterns Observed from Visual Analysis

From the journey duration distribution graph, most train routes are present in short and medium duration ranges. Long journey routes are comparatively fewer.

The station traffic analysis shows that most stations have low or medium traffic, while only major railway stations handle very high train movement.

The distance versus journey duration graph shows that journey duration generally increases as distance increases.

Some trains with similar distances have different journey durations because of factors such as number of stops, train speed, and waiting time.

Overall, distance and station traffic are important factors affecting railway journey patterns.
