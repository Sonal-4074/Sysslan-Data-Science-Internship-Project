```python
import pandas as pd

# Load dataset
df = pd.read_csv("Dataset1.csv")
```

```python
# Sort data properly
df = df.sort_values(["Train_No", "SN"])
```

```python
# Create total distance and number of stops for each train
train_features = df.groupby("Train_No").agg(
    Total_Distance=("Distance", "max"),
    Number_of_Stops=("Station_Name", "count")
).reset_index()
```

```python
# Display result
print(train_features.head(10))
```

```python
# Save output
train_features.to_csv("Train_Features.csv", index=False)
```
