```python
import pandas as pd

# Load dataset
df = pd.read_csv("Dataset1.csv")
```

```python
# Display time format before changes
print("Before formatting:")

print(df[["Arrival_time", "Departure_Time"]].head())
```

```python
# Convert arrival and departure time into standard time format

df["Arrival_time"] = pd.to_datetime(
    df["Arrival_time"],
    errors="coerce"
).dt.strftime("%H:%M:%S")

df["Departure_Time"] = pd.to_datetime(
    df["Departure_Time"],
    errors="coerce"
).dt.strftime("%H:%M:%S")
```

```python
# Display time after formatting

print("\nAfter formatting:")

print(df[["Arrival_time", "Departure_Time"]].head())
```

```python
df.to_csv("Time_Formatted_Data.csv", index=False)
```
