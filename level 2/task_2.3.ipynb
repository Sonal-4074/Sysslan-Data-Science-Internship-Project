```python
import pandas as pd

# Load dataset
df = pd.read_csv("Dataset1.csv")
```

```python
# Sort stations
df = df.sort_values(["Train_No", "SN"])
```

```python
# Convert time columns
df["Arrival_time"] = pd.to_datetime(
    df["Arrival_time"],
    format="%H:%M:%S"
)

df["Departure_Time"] = pd.to_datetime(
    df["Departure_Time"],
    format="%H:%M:%S"
)
```

```python
# Get start and end time of each train
journey = df.groupby("Train_No").agg(
    Start_Time=("Departure_Time", "first"),
    End_Time=("Arrival_time", "last")
).reset_index()
```

```python
# Handle next day arrival
journey.loc[
    journey["End_Time"] < journey["Start_Time"],
    "End_Time"
] += pd.Timedelta(days=1)
```

```python
# Calculate duration in hours
journey["Journey_Duration_Hours"] = (
    journey["End_Time"] - journey["Start_Time"]
).dt.total_seconds() / 3600
```

```python
journey["Journey_Duration_Hours"] = journey["Journey_Duration_Hours"].round(2)
```

```python
# Remove dates and keep only time
journey["Start_Time"] = journey["Start_Time"].dt.strftime("%H:%M:%S")
journey["End_Time"] = journey["End_Time"].dt.strftime("%H:%M:%S")
```

```python
# Display result
print(journey.head(10))
```

```python
journey.to_csv("Journey_Duration.csv", index=False)
```
